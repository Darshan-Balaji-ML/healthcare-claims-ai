# Data Preparation and Feature Engineering
## Week 3 — Cleaning and Feature Engineering
This notebook prepares the synthetic claims dataset for modeling. Steps include:
- Loading and inspecting the data
- Handling missing values
- Encoding categorical variables
- Engineering 7 domain-informed features
- Preparing NLP text features for the ICD classifier
- Train-test split

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import os

df = pd.read_csv('data/synthetic/synthetic_claims.csv')
print(df.shape)
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumn types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nBasic statistics:")
df.describe()
```

## Data Quality Check
- No missing values in any column
- No duplicate rows
- 50,000 claims loaded correctly
- cpt_code loaded as int64 — needs to be converted to string 
  since CPT codes are categorical identifiers, not numbers
- service_date loaded as object — needs to be converted to datetime

```python
# Fix dtypes
df['cpt_code'] = df['cpt_code'].astype(str)
df['service_date'] = pd.to_datetime(df['service_date'])

print("cpt_code dtype:", df['cpt_code'].dtype)
print("service_date dtype:", df['service_date'].dtype)
```

## Feature Engineering
Creating 7 domain-informed features based on patterns identified in EDA 
and healthcare billing domain knowledge.

```python
# Feature 1: billed_to_median_ratio
# Upcoding signal — how much more than the median is this claim billing?
cpt_medians = df.groupby('cpt_code')['billed_amount'].median()
df['billed_to_median_ratio'] = df.apply(
    lambda row: row['billed_amount'] / cpt_medians[row['cpt_code']], 
    axis=1
)

print("billed_to_median_ratio stats:")
print(df['billed_to_median_ratio'].describe())
```

```python
# Feature 2: filing_delay_bucket
# Bins days_to_submission into risk categories
bins = [0, 7, 30, 60, 91]
labels = ['0-7', '8-30', '31-60', '61+']

df['filing_delay_bucket'] = pd.cut(
    df['days_to_submission'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

print("Filing delay distribution:")
print(df['filing_delay_bucket'].value_counts().sort_index())
```

```python
# Feature 3: patient_age_group
age_bins = [0, 30, 50, 65, 80, 91]
age_labels = ['<30', '30-50', '50-65', '65-80', '80+']

df['patient_age_group'] = pd.cut(
    df['patient_age'],
    bins=age_bins,
    labels=age_labels,
    include_lowest=True
)

print("Age group distribution:")
print(df['patient_age_group'].value_counts().sort_index())
```

```python
# Feature 4: high_cost_outlier
# Flag claims above the 95th percentile for their CPT code
percentile_95 = df.groupby('cpt_code')['billed_amount'].transform(
    lambda x: x.quantile(0.95)
)

df['high_cost_outlier'] = (df['billed_amount'] > percentile_95).astype(int)

print("High cost outlier distribution:")
print(df['high_cost_outlier'].value_counts())
print(f"\n{df['high_cost_outlier'].mean():.2%} of claims flagged as high cost outliers")
```

```python
# Feature 5: duplicate_flag
# Flag claims where same patient age + CPT code + service date appears more than once
df['duplicate_flag'] = df.duplicated(
    subset=['patient_age', 'cpt_code', 'service_date'], 
    keep=False
).astype(int)

print("Duplicate flag distribution:")
print(df['duplicate_flag'].value_counts())
print(f"\n{df['duplicate_flag'].mean():.2%} of claims flagged as duplicates")
```

```python
# Feature 6: provider_denial_rate
# Historical denial rate per specialty
# (using specialty as proxy since we don't have individual provider IDs)
specialty_denial_rate = df.groupby('provider_specialty')['is_denied'].mean()
df['provider_denial_rate'] = df['provider_specialty'].map(specialty_denial_rate)

print("Provider denial rate stats:")
print(df['provider_denial_rate'].describe())
```

```python
# Feature 7: icd_chapter
# Extract ICD chapter from ICD code for grouping
icd_chapter_map = {
    'E11.9': 'Chapter IV - Endocrine/Metabolic',
    'I10': 'Chapter IX - Circulatory',
    'M54.5': 'Chapter XIII - Musculoskeletal',
    'J18.9': 'Chapter X - Respiratory',
    'F32.1': 'Chapter V - Mental Health',
    'I21.0': 'Chapter IX - Circulatory',
    'J44.1': 'Chapter X - Respiratory',
    'M17.11': 'Chapter XIII - Musculoskeletal',
    'Z23': 'Chapter XXI - Preventive',
    'N18.3': 'Chapter XIV - Genitourinary',
    'F41.1': 'Chapter V - Mental Health',
    'I50.9': 'Chapter IX - Circulatory',
    'E78.5': 'Chapter IV - Endocrine/Metabolic',
    'M79.3': 'Chapter XIII - Musculoskeletal',
    'G43.909': 'Chapter VI - Nervous System',
    'K21.0': 'Chapter XI - Digestive',
    'L40.0': 'Chapter XII - Skin',
    'F20.9': 'Chapter V - Mental Health',
    'C34.10': 'Chapter II - Neoplasms',
    'S72.001A': 'Chapter XIX - Injury'
}

df['icd_chapter'] = df['icd_code'].map(icd_chapter_map)

print("ICD chapter distribution:")
print(df['icd_chapter'].value_counts())
```

## Feature Engineering Summary

| Feature | Type | What it captures |
|---------|------|-----------------|
| `billed_to_median_ratio` | Float | Upcoding signal — how much above CPT median is this claim |
| `filing_delay_bucket` | Category | Late filing risk — bins days to submission into 4 groups |
| `patient_age_group` | Category | Age-based coverage rules — bins patient age into 5 groups |
| `high_cost_outlier` | Binary | Extreme billing flag — top 5% billed amount for CPT code |
| `duplicate_flag` | Binary | Duplicate billing detection — same patient/CPT/date |
| `provider_denial_rate` | Float | Provider-level risk — historical denial rate by specialty |
| `icd_chapter` | Category | Disease category grouping for fairness analysis |
